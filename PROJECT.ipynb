Guess the number

```python
import random

target = random.randint(1,100)

while True:
    userChoice =input("guess th number or Quiet:")
    if(userChoice == "Quiet"):
        break

    userChoice = int(userChoice)
    if(userChoice == target):
        print("success : correct guess!!")
        break
    elif(userChoice < target):
        print("your number are to small.take a bigger gusse....")
    else:
        print("your number are to big.take a smaller gusse....")


    print("------GAME OVER-----")    
```
